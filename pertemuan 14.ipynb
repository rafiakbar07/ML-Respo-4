```python
import pandas as pd
path = "asthma_disease_data.csv"
df = pd.read_csv(path)
print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
df = df.dropna() 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Distribusi Usia Pasien")
plt.xlabel("Usia")
plt.show()

```

```python
numerical_cols = ['Age', 'BMI', 'PollutionExposure', 'PollenExposure', 'DustExposure', 'LungFunctionFEV1', 'LungFunctionFVC']
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Matriks Korelasi Fitur Numerik")
plt.show()
```
